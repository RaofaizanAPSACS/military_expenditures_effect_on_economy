```python
# Load the csv file into dataframe object.
me = pd.read_csv("militaryExp.csv",skiprows=[0,1,2,3])
```

```python
me.head()
```

```python
# Select the only pakistan in all countries.
index= 0
for i, val in enumerate(me["Country Name"]):
  if val == "Pakistan":
    index = i
pakME = me.iloc[index:index+1]   
```

```python
pakME.columns
```

```python
# Drop unusefull column names.
pakME = pakME.drop(["Country Name","Country Code","Indicator Name","Indicator Code","2021","Unnamed: 66"],axis=1)
pakME.head()
```

```python
# Store the military expenditues by year in a list.
values = []
for i in pakME.columns:
  values.append(pakME[i])
```

```python
# Convert the list into a numpy array.
values = np.array(values)
values.astype("int64")
```

```python
# Select 30 years data.
columns = pakME.columns
col = columns[0:30]
val = values[0:30]
```

```python
# Visualize the trend of military expenditures.
plt.figure(figsize = (30,15))
plt.plot(col,val/1000000000,'go--')
plt.title('Pakistan')
plt.xlabel('years')
plt.ylabel('Military Spending in US Billion($)')
plt.savefig("MilitaryExp.png")
plt.show()
```

```python
# Select the latest 30 years span.
col = columns[30:]
val = values[30:]
```

```python
plt.figure(figsize = (30,15))
plt.plot(col,val/1000000000,'go--')
plt.title('Pakistan')
plt.xlabel('years')
plt.ylabel('Military Spending in US Billion($)')
plt.savefig("MilitaryExp(1).png")
plt.show()
```

```python
plt.figure(figsize = (40,15))
plt.plot(columns,values/1000000000,'go--')
plt.title('Pakistan')
plt.xlabel('years')
plt.ylabel('Military Spending in US Billion($)')
plt.savefig("MilitaryExpFull.png")
plt.show()
```

```python
df = pd.DataFrame(columns,columns=["years"])
df["Amount"] = values/1000000000
df.head()
```

```python
# Save the updated dataframe to csv file.
df.to_csv("militaryExpenditure.csv")
```
